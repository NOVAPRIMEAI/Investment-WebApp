## APPLE

```python
import pandas as pd

df = pd.read_csv("data/AAPL.csv", parse_dates=['Date'])
df.head()
```

```python
df['Month'] = df['Date'].dt.strftime("%Y-%m")
df_mon = df.groupby(['Month']).mean()
df_mon.head()
```

```python
import matplotlib.pyplot as plt

#Gives the plotted graph a size -> 25 length and 7 height
plt.figure(figsize=(25,7))

#plot function takes a list of x and y coordinates
#x coordinates are the month values -> index of the dataframe
#y coordinates are the cost for each month value
plt.plot(df_mon.index, df_mon["High"])
```
