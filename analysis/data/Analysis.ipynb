```python
import yfinance as yf

df = yf.download("NKE", start="2018-11-01", end="2020-11-1", interval="1d")
df.head()
```

```python
msft = yf.Ticker("MSFT")
```

```python
msft.dividends
```

```python
msft.dividends.plot(figsize=(14, 7))
```

```python
import plotly.graph_objects as go
fig = go.Figure(
    data=go.Ohlc(
        x=df.index,
        open=df["Open"],
        high=df["High"],
        low=df["Low"],
        close=df["Close"],
    )
)
fig.show()
```
